```python
input_folder_path = "/home/ajarrah/PhD_Thesis/chapter_2/h5ad_data_labeled/"
output_folder_path = "/home/ajarrah/PhD_Thesis/chapter_2/h5ad_data_labeled_separated/"
file_name_list_csv_output_path = "/home/ajarrah/PhD_Thesis/chapter_2/csv_data/file_name_list_hpc.csv"
hpc_column = "hpc"  # Column name to filter
```

```python
import os
import anndata as ad
import pandas as pd

```

```python
# Create output folder if it doesn't exist
os.makedirs(output_folder_path, exist_ok=True)

# ==== PROCESS ALL FILES ====
for file_name in os.listdir(input_folder_path):
    if file_name.endswith(".h5ad"):
        file_path = os.path.join(input_folder_path, file_name)
        print(f"Processing: {file_path}")
        
        # Load AnnData
        adata = ad.read_h5ad(file_path)
        
        # Check if 'hpc' column exists in obs
        if hpc_column not in adata.obs.columns:
            print(f"  Skipped: Column '{hpc_column}' not found in {file_name}")
            continue
        
        # Filter rows where hpc == True (or non-zero)
        mask = adata.obs[hpc_column].astype(bool)
        filtered_adata = adata[mask].copy()
        
        print(f"  Kept {filtered_adata.n_obs} pixels labeled as '{hpc_column}'")
        
        # Save filtered AnnData if not empty
        if filtered_adata.n_obs > 0:
            output_path = os.path.join(output_folder_path, f"filtered_{file_name}")
            filtered_adata.write(output_path)
            print(f"  Saved to: {output_path}")
        else:
            print(f"  No pixels with '{hpc_column}' == True found in {file_name}")
```

```python
file_name_list = []
for file_name in os.listdir(output_folder_path):
    if file_name.endswith(".h5ad"):
        file_name_list.append(file_name)
file_name_list.sort()
file_name_list
```

```python
pd.DataFrame(file_name_list, columns=["file_name"]).to_csv(file_name_list_csv_output_path, index=False)
```
