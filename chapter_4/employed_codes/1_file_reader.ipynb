```python
input_folder = "/opt/Data/RNA Seq 2025"
file_name_list_csv_output_path = "/home/ajarrah/PhD_Thesis/chapter_3/csv_data/file_name_list.csv"
h5ad_output_path = "/home/ajarrah/PhD_Thesis/chapter_3/h5ad_data/"
```

```python
import os
import pandas as pd
import scanpy as sc
import numpy as np
```

```python
file_name_list = []
for file_name in os.listdir(input_folder):
    file_name_list.append(file_name)
file_name_list.sort()
file_name_list
```

```python
pd.DataFrame(file_name_list, columns=["file_name"]).to_csv(file_name_list_csv_output_path, index=False)
```

```python

# Group abbreviation and main code mapping
group_map = {
    ('Young', 'Control'): ('YC', 1),
    ('Young', 'AD'): ('YAD', 2),
    ('Aged', 'Control'): ('AC', 3),
    ('Aged', 'AD'): ('AAD', 4)
}

# Letter-to-number mapping for sub-index
letter_map = {'A': 1, 'B': 2, 'C': 3, 'D': 4}

# Parse data
parsed_data = []
for sample in file_name_list:
    parts = sample.split('_')
    letter = parts[0][0]  # A, B, C, D
    age = parts[1]
    disease = parts[2]
    
    group_abbrev, main_num = group_map.get((age, disease), ('Unknown', 0))
    sub_num = letter_map.get(letter, 0)
    code = f"{main_num}-{sub_num}"  # e.g., 1-1, 2-2
    adata = sc.read_visium("/opt/Data/RNA Seq 2025/"+sample+"/outs/")
    adata.obs["Original_Name"] = sample
    adata.obs["Sample_Code"] = code
    adata.obs["Group"] = group_abbrev
    adata.var_names_make_unique()
    adata.write(h5ad_output_path+sample+".h5ad")
```
