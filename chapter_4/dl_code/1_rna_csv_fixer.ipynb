```python
import numpy as np
import pandas as pd

RNA_TISSUE_POSITIONS_CSV = "/home/ajarrah/PhD_Thesis/chapter_4/tissue_positions/tissue_positions.csv"
rna_spot_spacing = 100  # micrometers
```

```python

tissue_positions = pd.read_csv(RNA_TISSUE_POSITIONS_CSV, usecols=["barcode","array_row","array_col", "pxl_row_in_fullres", "pxl_col_in_fullres"])

# Shift full-resolution pixel coordinates to start at zero
min_row = tissue_positions["pxl_row_in_fullres"].min()
min_col = tissue_positions["pxl_col_in_fullres"].min()

tissue_positions["pxl_row_in_fullres_s"] = tissue_positions["pxl_row_in_fullres"] - min_row
tissue_positions["pxl_col_in_fullres_s"] = tissue_positions["pxl_col_in_fullres"] - min_col
coords_um = []
for _, row in tissue_positions.iterrows():
    r, c = row['array_row'], row['array_col']
    y = r * (np.sqrt(3)/2 * rna_spot_spacing)
    x = c * rna_spot_spacing/2
    coords_um.append((x, y))

coords_um = np.array(coords_um)
tissue_positions['x_um'] = coords_um[:, 0]
tissue_positions['y_um'] = coords_um[:, 1]

```
