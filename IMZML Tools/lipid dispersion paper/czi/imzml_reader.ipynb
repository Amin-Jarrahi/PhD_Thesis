```python
from czifile import CziFile
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Path to your .czi file
czi_path = '2025_06_17__7306.czi'

# Open and read the CZI file
with CziFile(czi_path) as czi:
    image_data = czi.asarray()  # Returns a numpy array
    print(f"Image shape: {image_data.shape}")

# Show a 2D slice from the image
# Adjust indexing based on shape: (T, C, Z, Y, X, 1) or similar
plt.imshow(image_data[0, 0, 0, :, :, 0], cmap='coolwarm')
plt.title("CZI Image Slice")
plt.axis('off')
plt.show()
```
